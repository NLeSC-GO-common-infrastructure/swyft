```python
import numpy as np
import pylab as plt
import swyft
import h5py
import torch
from scipy import ndimage
%load_ext autoreload
%autoreload 2
```

```python
DEVICE = 'cuda:0'
NDIM = 2 # number of parameters
NBATCH = 4  # minibatch size
NTRAIN = 1000
NEPOCHS = NTRAIN // (NTRAIN // NBATCH) # training epochs
NSAMPLES = 1000 # traing batch size
NWORKERS = 1 # number of cpu workers for dataloader
```

```python
class Data(torch.utils.data.Dataset):
    def __init__(self, N = None):
        super().__init__()
        self.data = h5py.File("/home/ubuntu/volume1/pyrofit_lensing/yaml/methods_paper/cw/train_v2.hdf5", "r")
        self.N = len(self.data['mu']) if N is None else N
        self.noise = None

    def __len__(self):
        return self.N
    
    def set_noise(self, noise):
        self.noise = noise

    def __getitem__(self, idx, noise = None):
        if isinstance(idx, int):
            idx = slice(idx, idx+1)
            flag = True
        else:
            flag = False
        x = self.data['mu'][idx]
        x -= self.data['mu'][0]
        x_sub = self.data['main']['sub']['x_sub'][idx]
        y_sub = self.data['main']['sub']['y_sub'][idx]
        i_sub = [m.argsort()[-100:] for m in self.data['main']['sub']['m_sub'][idx]]
        m_sub = [m[i] for m, i in zip(self.data['main']['sub']['m_sub'][idx], i_sub)]
        x_sub = [x[i] for x, i in zip(self.data['main']['sub']['x_sub'][idx], i_sub)]
        y_sub = [y[i] for y, i in zip(self.data['main']['sub']['y_sub'][idx], i_sub)]
        x = torch.tensor(x).float()
        z = torch.tensor([x_sub, y_sub]).float()
        z = torch.transpose(z, 0, 1)
        z = (z+2)/4
        if self.noise is None:
            noise = torch.rand(1)
        else:
            noise = self.noise
        x += torch.randn_like(x)*noise
        #x = x*z[:,0]
        #x[:, :200, :] = z[:,0,0]
        #x[:, 200:, :] = z[:,1,0]
        #x = z
        if flag:
            x = x[0]
            z = z[0]
        return {"x":x, "z":z}
```

```python
# Generate training data
dataset = Data(N = 10000)
N = len(dataset)
N_train = 9500
N_valid = 500
dataset_train, dataset_valid = torch.utils.data.random_split(dataset, [N_train, N_valid])

data_train_loader = torch.utils.data.DataLoader(dataset_train, batch_size=NBATCH, num_workers=NWORKERS, pin_memory=True, drop_last=True)
data_valid_loader = torch.utils.data.DataLoader(dataset_valid, batch_size=NBATCH, num_workers=NWORKERS, pin_memory=True, drop_last=True)
```

```python
# Convolutional network as HEAD of inference network

class Head(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = torch.nn.Conv2d(1, 64, 5, padding = 2)
        self.conv2 = torch.nn.Conv2d(64, 64, 5, padding =  2)
        self.conv3 = torch.nn.Conv2d(64, 64, 5, padding = 2)
        self.conv4 = torch.nn.Conv2d(64, 64, 5, padding = 2)
        self.conv5 = torch.nn.Conv2d(64, 64, 5, padding = 2)
        self.conv6 = torch.nn.Conv2d(64, 8, 5)
        #self.apool = torch.nn.AvgPool2d(1)
        self.mpool = torch.nn.MaxPool2d(2)
        
    def forward(self, x):
        nbatch = len(x)
        
        x = x.unsqueeze(1)
        #x = self.conv1(x)
        #x = self.mpool(x)
        #x = self.mpool(x)
        #x = self.mpool(x)
        #x = self.apool(x)
        x = self.conv1(x)
        x = torch.relu(x)
        x = self.mpool(x)
        x = self.conv2(x)
        x = self.mpool(x)
        x = torch.relu(x)
        x = self.conv3(x)
        x = self.mpool(x)
        x = torch.relu(x)
        x = self.conv4(x)
        x = self.mpool(x)
        x = torch.relu(x)
        x = self.conv5(x)
        x = self.mpool(x)
        x = torch.relu(x)
        x = self.conv6(x)
        x = self.mpool(x)
        x = torch.relu(x)
        x = x.view(nbatch, -1)

        return x
```

```python
# Get number of features by evaluating Head network once on target data

x0 = dataset[0:1]['x']
ydim = Head()(x0).shape[1]  # Number of output features per image
print("Number of features:", ydim)
```

```python
dataset.set_noise(None)
```

```python
# Instantiate network

net = swyft.Network(xdim = ydim, zdim = NDIM, head = Head().to(DEVICE)).to(DEVICE)
train_loss, valid_loss = [], []
```

```python
# Train!

for nepochs, lr, noise in [
        [4, 1e-4, 0.01],
        [4, 1e-4, 0.1],
        [4, 1e-4, 0.2],
        [4, 1e-4, 0.4],
        [4, 1e-4, 0.6],
        [4, 1e-4, 0.8],
        [4, 1e-4, 1.0],
        [4, 1e-4, 1.0]]:
    noise = 0.0001
    lr = 3e-4
    dataset.set_noise(noise)
    nepochs = 1
    tl, vl, sd = swyft.train(net, data_train_loader, data_valid_loader, max_epochs = nepochs, early_stopping_patience = 20, lr = lr, device=DEVICE)
    vl_minimum = min(vl)
    vl_min_idx = vl.index(vl_minimum)
    train_loss.append(tl[:vl_min_idx + 1])
    valid_loss.append(vl[:vl_min_idx + 1])
    print(train_loss)
    net.load_state_dict(sd)
```
